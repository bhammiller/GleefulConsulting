```python
import pandas as pd
import statsmodels.formula.api as sm

Location="GleeClubData.xlsx"
df=pd.read_excel(Location)

df.head()
```

```python
df.corr()
```

```python
df2=df
headers=list(df2.columns.values)
headers[0]='team_size'
headers[1]='mf_ratio_team'
headers[2]='school_enrollment'
headers[3]='mf_ratio_school'
headers[4]='district_pop'
headers[5]='practices_per_month'
headers[6]='comp_before_reg'
headers[7]='reg_score'
df2.columns=headers
df2.head()
```

```python
result = sm.ols(formula='reg_score ~ team_size + mf_ratio_team + school_enrollment + school_enrollment + mf_ratio_school + district_pop + practices_per_month + comp_before_reg', data=df).fit()
result.summary()
```

```python
result = sm.ols(formula='reg_score ~ team_size + school_enrollment + school_enrollment + district_pop + practices_per_month + comp_before_reg -1', data=df).fit()
result.summary()
```
